```python
import matplotlib.pyplot as plt
import math
import numpy as np
```

```python
time_lfu = [0.141138, 0.519754, 0.351113, 0.492665, 1.6973,
            0.01193, 0.012693, 0.299197, 0.34217]
hits_lfu = [89999, 113501, 298999, 389999, 79830,
            2089, 742, 3946, 100314]
```

```python
p1 = np.polyfit(time_lfu, hits_lfu, 1)  # вычисление коэффициентов многочлена
y_approx_lfu = np.polyval(p1, time_lfu)
p1
```

```python
plt.xlabel('Time')  # Подпись для оси х
plt.ylabel('Hits')  # Подпись для оси y
plt.title('LFUCache')  # Название
plt.scatter(time_lfu, hits_lfu, color='blue', label='LFU Points')
plt.plot(time_lfu, y_approx_lfu, color='red', label='LFU Approx')

plt.grid(True)  # Включение основной сетки
plt.legend()  # Показ легенды
plt.show()
```

```python
time_perfect = [0.342655, 0.793287, 0.743219, 1.21592, 1.6233, 0.01535, 
                0.023967, 0.024173, 0.316135, 0.598922]
hits_perfect = [89999, 113501, 298999, 389999, 124450, 3145, 
                714, 1363, 26237, 130814]
```

```python
plt.xlabel('Time')  # Подпись для оси х
plt.ylabel('Hits')  # Подпись для оси y
plt.title('Compare caches')  # Название
plt.scatter(time_perfect, hits_perfect, color='pink', label='Perfect Points')
plt.plot(time_perfect, y_approx_perfect, color='red', label='Perfect Approx')
plt.scatter(time_lfu, hits_lfu, color='lightblue', label='LFU Points')
plt.plot(time_lfu, y_approx_lfu, color='blue', label='LFU Approx')

plt.grid(True)  # Включение основной сетки
plt.legend()  # Показ легенды
plt.show()
```
